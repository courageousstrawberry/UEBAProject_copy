```python
import json

# Load the JSON data
with open("data.json", "r") as f:
    data = json.load(f)
```

```python
# Check what kind of structure we're working with
if isinstance(data, dict):
    print("Top-level keys:", data.keys())

    # Extract the list under the "Records" key
    records = data["Records"]

    # Sanity check
    print(f"Type of 'Records': {type(records)}")
    print(f"Number of records: {len(records)}")

    # Show a sample record to understand structure
    import json
    print(json.dumps(records[0], indent=2))

elif isinstance(data, list):
    print(f"Top-level is a list with {len(data)} records.")
    print("First item preview:")
    print(json.dumps(data[0], indent=2))


```

```python
import pandas as pd

# Flatten the list of records into a DataFrame
df = pd.json_normalize(records)

# Preview the flattened table
df.head()

```
